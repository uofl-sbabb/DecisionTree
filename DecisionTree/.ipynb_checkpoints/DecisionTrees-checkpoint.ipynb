# Decision Trees:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
geopolymer_df = pd.read_csv("Artificial Intelligence-Based Life Cycle Assessment for Sustainable Geopolymer Concrete - Copy.csv")
geopolymer_df
```

```python
geopolymer_df = geopolymer_df.rename(columns={'Unnamed: 9': 'Comprehensive Strength 7 days(psi)', 'Unnamed: 10': 'Comprehensive Strength 28 days(psi)', 'Unnamed: 11': '170 degrees for 24 hours(psi)'})
geopolymer_df
```

```python
geopolymer_df.drop(geopolymer_df.index[0], inplace=True) 
geopolymer_df
```

# Input Values:

```python
geopolymer_df.head()
```

```python
mask_ist = geopolymer_df['IST(min)'].isna()
geopolymer_df.loc[mask_ist, 'IST(min)'] = np.random.uniform(low=30, high=120, size=mask_ist.sum())
# need to change the float type to int to use Decision Tree Classifier
geopolymer_df['IST(min)'] = geopolymer_df['IST(min)'].astype('int')

mask_fst = geopolymer_df['FST(min)'].isna()
geopolymer_df.loc[mask_fst, 'FST(min)'] = np.random.uniform(low=60, high=240, size=mask_fst.sum())
geopolymer_df['FST(min)'] = geopolymer_df['FST(min)'].astype('int')

mask_low_cs = geopolymer_df['Comprehensive Strength'].isna()
geopolymer_df.loc[mask_low_cs, 'Comprehensive Strength'] = np.random.uniform(low=400, high=3000, size=mask_low_cs.sum())
geopolymer_df['Comprehensive Strength'] = geopolymer_df['Comprehensive Strength'].astype('int')

mask_medium_cs = geopolymer_df['Comprehensive Strength 7 days(psi)'].isna()
geopolymer_df.loc[mask_medium_cs, 'Comprehensive Strength 7 days(psi)'] = np.random.uniform(low=1200, high=6000, size=mask_medium_cs.sum())
geopolymer_df['Comprehensive Strength 7 days(psi)'] = geopolymer_df['Comprehensive Strength 7 days(psi)'].astype('int')


mask_high_cs = geopolymer_df['Comprehensive Strength 28 days(psi)'].isna()
geopolymer_df.loc[mask_high_cs, 'Comprehensive Strength 28 days(psi)'] = np.random.uniform(low=2500, high=8000, size=mask_high_cs.sum())
geopolymer_df['Comprehensive Strength 28 days(psi)'] = geopolymer_df['Comprehensive Strength 28 days(psi)'].astype('int')


mask_170_degrees = geopolymer_df['170 degrees for 24 hours(psi)'].isna()
geopolymer_df.loc[mask_170_degrees, '170 degrees for 24 hours(psi)'] = np.random.uniform(low=1200, high=6500, size=mask_170_degrees.sum())
geopolymer_df['170 degrees for 24 hours(psi)'] = geopolymer_df['170 degrees for 24 hours(psi)'].astype('int')


mask_fs = geopolymer_df['FS'].isna()
geopolymer_df.loc[mask_fs, 'FS'] = np.random.uniform(low=500, high=1800, size=mask_fs.sum())
geopolymer_df['FS'] = geopolymer_df['FS'].astype('int')


geopolymer_df.head()
```

```python
geopolymer_df.describe()
```

# Train and Test Data:

```python
import sklearn
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# get X (input) and y (output) values
X = geopolymer_df.iloc[:, 1:6]
y = geopolymer_df.iloc[:, 6:13]
y_IST = geopolymer_df.iloc[:, 6]
y_FST = geopolymer_df.iloc[:, 7]
y_CS3 = geopolymer_df.iloc[:, 8]
y_CS7 = geopolymer_df.iloc[:, 9]
y_CS28 = geopolymer_df.iloc[:, 10]
y_170degrees = geopolymer_df.iloc[:, 11]
y_FS = geopolymer_df.iloc[:, 12]
```

```python
# train and test the data
X_train_IST, X_test_IST, y_train_IST, y_test_IST = train_test_split(X, y_IST, random_state=42, test_size=0.2)

X_train_FST, X_test_FST, y_train_FST, y_test_FST = train_test_split(X, y_FST, random_state=42, test_size=0.2)

X_train_CS3, X_test_CS3, y_train_CS3, y_test_CS3 = train_test_split(X, y_CS3, random_state=42, test_size=0.2)

X_train_CS7, X_test_CS7, y_train_CS7, y_test_CS7 = train_test_split(X, y_CS7, random_state=42, test_size=0.2)

X_train_CS28, X_test_CS28, y_train_CS28, y_test_CS28 = train_test_split(X, y_CS28, random_state=42, test_size=0.2)

X_train_170degrees, X_test_170degrees, y_train_170degrees, y_test_170degrees = train_test_split(X, y_170degrees, random_state=42, test_size=0.2)

X_train_FS, X_test_FS, y_train_FS, y_test_FS = train_test_split(X, y_FS, random_state=42, test_size=0.2)
```

```python
# Decision Tree fit
dt_IST = tree.DecisionTreeClassifier()
dt_IST = dt_IST.fit(X_train_IST, y_train_IST)

dt_FST = tree.DecisionTreeClassifier()
dt_FST = dt_FST.fit(X_train_FST, y_train_FST)

dt_CS3 = tree.DecisionTreeClassifier()
dt_CS3 = dt_CS3.fit(X_train_CS3, y_train_CS3)

dt_CS7 = tree.DecisionTreeClassifier()
dt_CS7 = dt_CS7.fit(X_train_CS7, y_train_CS7)

dt_CS28 = tree.DecisionTreeClassifier()
dt_CS28 = dt_CS28.fit(X_train_CS28, y_train_CS28)

dt_170degrees = tree.DecisionTreeClassifier()
dt_170degrees = dt_170degrees.fit(X_train_170degrees, y_train_170degrees)

dt_FS = tree.DecisionTreeClassifier()
dt_FS = dt_FS.fit(X_train_FS, y_train_FS)
```

```python
# predictions
y_pred_IST = dt_IST.predict(X_test_IST)

y_pred_FST = dt_FST.predict(X_test_FST)

y_pred_CS3 = dt_CS3.predict(X_test_CS3)

y_pred_CS7 = dt_CS7.predict(X_test_CS7)

y_pred_CS28 = dt_CS28.predict(X_test_CS28)

y_pred_170degrees = dt_170degrees.predict(X_test_170degrees)

y_pred_FS = dt_FS.predict(X_test_FS)
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test_IST, y_pred_IST))
```

```python
# Confusion Matrix FST
print(confusion_matrix(y_test_FST, y_pred_FST))
```

```python
# Confusion Matrix Comprehensive Strength 3 days(psi)
print(confusion_matrix(y_test_CS3, y_pred_CS3))
```

```python
# Confusion Matrix Comprehensive Strength 7 days(psi)
print(confusion_matrix(y_test_CS7, y_pred_CS7))
```

```python
# Confusion Matrix Comprehensive Strength 28 days(psi)
print(confusion_matrix(y_test_CS28, y_pred_CS28))
```

```python
# Confusion Matrix 170 degrees 24 hours (psi)
print(confusion_matrix(y_test_170degrees, y_pred_170degrees))
```

```python
# Confusion Matrix Flexural Strength
print(confusion_matrix(y_test_FS, y_pred_FS))
```
