```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
geopolymer_df.fillna(value=0, inplace=True)
geopolymer_df
```

```python
geopolymer_df = geopolymer_df.rename(columns={'Unnamed: 9': 'Comprehensive Strength 7 days(psi)', 'Unnamed: 10': 'Comprehensive Strength 28 days(psi)', 'Unnamed: 11': '170 degrees for 24 hours(psi)'})
geopolymer_df.info()
```

```python
geopolymer_df.drop(geopolymer_df.index[0], inplace=True) 
geopolymer_df
```

```python
geopolymer_df.info()
```

```python
geopolymer_df['Comprehensive Strength'] = pd.to_numeric(geopolymer_df['Comprehensive Strength'], errors='coerce')
geopolymer_df['Comprehensive Strength 7 days(psi)'] = pd.to_numeric(geopolymer_df['Comprehensive Strength 7 days(psi)'], errors='coerce')
geopolymer_df['Comprehensive Strength 28 days(psi)'] = pd.to_numeric(geopolymer_df['Comprehensive Strength 28 days(psi)'], errors='coerce')
geopolymer_df['170 degrees for 24 hours(psi)'] = pd.to_numeric(geopolymer_df['170 degrees for 24 hours(psi)'], errors='coerce')
geopolymer_df
```

```python
IST = geopolymer_df['IST(min)']
FST = geopolymer_df['FST(min)']
CS = geopolymer_df['Comprehensive Strength']
CS_7days = geopolymer_df['Comprehensive Strength 7 days(psi)']
CS_28days = geopolymer_df['Comprehensive Strength 28 days(psi)']
degrees_170 = geopolymer_df['170 degrees for 24 hours(psi)']
FS = geopolymer_df['FS']
```

```python
import sklearn
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
geopolymer_df.head()
```

```python
X = geopolymer_df.iloc[:, 0:6]
y_IST = geopolymer_df.iloc[:, 6]
y_FST = geopolymer_df.iloc[:, 7]
y_CS3 = geopolymer_df.iloc[:, 8]
y_CS7 = geopolymer_df.iloc[:, 9]
y_CS28 = geopolymer_df.iloc[:, 10]
y_170degrees = geopolymer_df.iloc[:, 11]
y = geopolymer_df.iloc[:, 6:13]
```

- Get the X (Input values) and y (output values)

```python
# IST training and testing set
X_train_ist, X_test_ist, y_train_ist, y_test_ist = train_test_split(X, y_IST, random_state=42, test_size=0.2)
```

```python
# FST training and testing set
X_train_fst, X_test_fst, y_train_fst, y_test_fst = train_test_split(X, y_FST, random_state=42, test_size=0.2)
```

```python
# Comprehensive Strength of 3 days(psi) training and testing set
X_train_cs3, X_test_cs3, y_train_cs3, y_test_cs3 = train_test_split(X, y_CS3, random_state=42, test_size=0.2)
```

```python
# Comprehensive Strength of 7 days(psi) training and testing set
X_train_cs7, X_test_cs7, y_train_cs7, y_test_cs7 = train_test_split(X, y_CS7, random_state=42, test_size=0.2)
```

```python
# Comprehensive Strength of 28 days(psi) training and testing set
X_train_cs28, X_test_cs28, y_train_cs28, y_test_cs28 = train_test_split(X, y_CS28, random_state=42, test_size=0.2)
```

```python
# 170 degrees training and testing set
X_train_170, X_test_170, y_train_170, y_test_170 = train_test_split(X, y_170degrees, random_state=42, test_size=0.2)
```

```python
# All the output values training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

- Create the training and testing sets from the input (X) and output (y) of the data.

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

- Get shape of train and test of the input/output values.

```python
dt = DecisionTreeClassifier()
```

```python
dt.get_params()
```

```python
dt.fit(X_train, y_train)
```

- Create fit the decision tree classifier.

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor(
    max_depth=7,
    min_samples_split=7,
    random_state=42
)

dtr.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
y_pred = dt.predict(X_test)
```

```python
y_pred_dtr = dtr.predict(X_test)
```

```python
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

- MSE is low indicating (0) means the model is perfect.
- R2 score ranges from 0 to 1. A value of 0 means model is perfect.

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred_dtr))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```
